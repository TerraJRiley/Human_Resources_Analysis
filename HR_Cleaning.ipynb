# General DF Cleaning

```python
# Filling Missing Values in 'ManagerID' Column for Webster Butler
df["ManagerID"] = df["ManagerID"].fillna(40)
```

```python
# ChatGPT Code
# Updating EmpStatusID for employees "Terminated for Cause"
# (There were two that I suspect were misslabled, so we're fixing those)

# Identifying the indices of employees to update
indices_to_update = df[(df['EmpStatusID'] == 1) & (df['EmploymentStatus'] == 'Terminated for Cause')].index

# Updating EmpStatusID to 4 for these employees
df.loc[indices_to_update, 'EmpStatusID'] = 4

# Displaying the updated records for confirmation
updated_records = df.loc[indices_to_update, ['Employee_Name', 'EmpStatusID', 'EmploymentStatus']]
updated_records
```

```python
# Edited ChatGPT Code
# Updating the 'EmploymentStatus' for employees with an active status to include "Active_" followed by their EmpStatusID

# Filtering for employees with active status
active_employees = df[df['EmploymentStatus'] == 'Active']

# Updating 'EmploymentStatus' to "Active_" + corresponding EmpStatusID
df.loc[active_employees.index, 'EmploymentStatus'] = 'Active_' + df.loc[active_employees.index, 
                                                                        'EmpStatusID'].astype(str)

# Verifying the update with a sample
df[df['EmploymentStatus'].str.startswith('Active')][
    ['Employee_Name', 'EmpStatusID', 'EmploymentStatus']].head(10)
```

```python
# Edited ChatGPT Code
# Adjusting the code to replace only NaT (not a time) values in 'DateofTermination' with the current date
# This will handle cases where 'DateofTermination' is missing, instead of future dates

# Re-loading the dataset to reset the changes

# Converting 'DateofHire' and 'DateofTermination' to datetime objects
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'], errors='coerce')  # 'coerce' will set invalid parsing as NaT

# Replacing NaT in 'DateofTermination' with the current date
#current_date = datetime.now()
df['DateofTermination'].fillna(latest_date, inplace=True)

# Calculating the difference between 'DateofHire' and 'DateofTermination'
df['EmploymentDuration'] = (df['DateofTermination'] - df['DateofHire']).dt.days

# Verifying the changes with a sample
df[['Employee_Name', 'DateofHire', 'DateofTermination', 'EmploymentDuration']].head(5)
```

```python
def map_funct(cell_in):
    if cell_in == "MA":
        cell_out = 1
    else:
        cell_out = 0
    return cell_out
df['On_Location'] = df['State'].map(lambda x: map_funct(x))
df = df.drop(["State"], axis = 1)
```
