```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/HR_Cleaned.csv")
df.head()
```

```python
df["HispanicLatino"] = df["HispanicLatino"].str.capitalize().str.strip()
```

```python

# Function to calculate age
def calculate_age(birthdate, current_date):
    return current_date.year - birthdate.year - ((current_date.month, current_date.day) < (birthdate.month, birthdate.day))

# Function to calculate tenure
def calculate_tenure(hire_date, termination_date, current_date):
    # If the employee is still working, use current_date as the end date
    if pd.isna(termination_date):
        end_date = current_date
    else:
        end_date = termination_date
    return (end_date - hire_date).days

# Code to add Age and Tenure columns
def add_columns(df):
    # Convert date columns to datetime
    df['DOB'] = pd.to_datetime(df['DOB'])
    df['DateofHire'] = pd.to_datetime(df['DateofHire'])
    df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])

    # Define the current date
    current_date = datetime(2023, 1, 30)

    # Calculate Age
    df['Age'] = df['DOB'].apply(lambda x: calculate_age(x, current_date))

    # Calculate Tenure
    df['Tenure'] = df.apply(lambda x: calculate_tenure(x['DateofHire'], x['DateofTermination'], current_date), axis=1)

    return df

# Example usage:
df = add_columns(df)
df.head()
```

```python
# Create a side-by-side plot with a bar chart and a pie chart for marital status distribution


# Count the frequency of each marital status
marital_status_counts = df["MaritalDesc"].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart
sns.barplot(ax=axes[0], x=marital_status_counts.index, y=marital_status_counts.values, palette="viridis")
axes[0].set_title('Marital Status Distribution - Bar Chart')
axes[0].set_xlabel('Marital Status')
axes[0].set_ylabel('Number of Employees')
axes[0].tick_params(axis='x', rotation=45)

# Pie chart
axes[1].pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', startangle=140, 
            colors=sns.color_palette("viridis", len(marital_status_counts)))
axes[1].set_title('Marital Status Distribution - Pie Chart')

plt.tight_layout()
plt.show()
```

```python
# Count the frequency of each race
race_counts = df["RaceDesc"].value_counts()

# Create a side-by-side plot
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Bar chart with count labels
sns.barplot(ax=axes[0], x=race_counts.index, y=race_counts.values, palette="Set3")
axes[0].set_title('Race Distribution - Bar Chart')
axes[0].set_xlabel('Race')
axes[0].set_ylabel('Number of Employees')
axes[0].tick_params(axis='x', rotation=90)

# Adding the count above each bar
for p in axes[0].patches:
    axes[0].annotate(f"{int(p.get_height())}", (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=10, color='gray', xytext=(0, 5),
                     textcoords='offset points')

# Pie chart with adjusted labels
pie_colors = sns.color_palette("Set3", len(race_counts))
wedges, texts, autotexts = axes[1].pie(race_counts, labels=None, autopct='%1.1f%%', startangle=140, colors=pie_colors)
axes[1].set_title('Race Distribution - Pie Chart')

# Adjusting the legend
race_percentage = [f"{label}: {pct}%" for label, pct in zip(race_counts.index, (race_counts/race_counts.sum()*100).round(1))]
axes[1].legend(wedges, race_percentage, title="Race and Percentage", loc='lower center', bbox_to_anchor=(0.5, -0.5), ncol=1)

plt.tight_layout()
plt.show()

# Usage:
# generate_race_distribution_charts(your_dataframe, 'RaceDesc')
```

```python
# Define the specific colors for F and M from the transgender flag
gender_colors = {"F": "#F7A8B8", "M": "#55CDFC"}

# Count the frequency of each gender
gender_counts = df["Sex"].value_counts()

# Create custom color palette based on gender values
custom_palette = [gender_colors.get(gender, "#000000") for gender in gender_counts.index]

# Create a side-by-side plot with a bar chart and a pie chart for gender distribution
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart
sns.barplot(ax=axes[0], x=gender_counts.index, y=gender_counts.values, palette=custom_palette)
axes[0].set_title('Gender Distribution - Bar Chart')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Number of Employees')

# Pie chart
axes[1].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=custom_palette)
axes[1].set_title('Gender Distribution - Pie Chart')

plt.tight_layout()
plt.show()
```

```python
# Adjusting the visualizations to place them side-by-side and using a single color spectrum for both heatmaps

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# Heatmap for Race and FromDiversityJobFair
sns.heatmap((pd.crosstab(df['RaceDesc'], df['FromDiversityJobFairID'], 
                         normalize='all') * 100), annot=True, cmap='Blues', fmt='.1f', ax=axes[0])
axes[0].set_title('Percentage of Total for Race and FromDiversityJobFair Cross-Section')
axes[0].set_ylabel('Race')
axes[0].set_xlabel('From Diversity Job Fair (1 = Yes, 0 = No)')

# Heatmap for Race and HispanicLatino
sns.heatmap((pd.crosstab(df['RaceDesc'], df['HispanicLatino'], 
                        normalize='all') * 100), annot=True, cmap='Blues', fmt='.1f', ax=axes[1])
axes[1].set_title('Percentage of Total for Race and HispanicLatino Cross-Section')
axes[1].set_ylabel('Race')
axes[1].set_xlabel('Hispanic/Latino')

plt.tight_layout()
plt.show()
```

# Function to clean text data by converting to lowercase and stripping whitespace
def clean_text_data(df, column_name):
    df[column_name] = df[column_name].str.lower().str.strip()
    return df

# Function to clean gender data by stripping whitespace from the gender values
def clean_gender_data(df, gender_column):
    df[gender_column] = df[gender_column].str.strip()
    return df
